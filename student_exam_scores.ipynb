```python
plt.figure(figsize=(10, 6))
plt.scatter(df['hours_studied'], df['exam_score'])
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs. Exam Score')
plt.grid(True)
plt.show()
```

The scatter plot shows the relationship between "Hours Studied" and "Exam Score". Based on the plot,there is a positive correlation between the two variables. As the number of hours studied increases, the exam score tends to increase as well. The points are not perfectly aligned, which suggests that other factors also influence the exam score, but there is a general trend showing that more study time is associated with higher exam scores.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sleep_hours'], df['exam_score'])
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.title('Sleep Hours vs. Exam Score')
plt.grid(True)
```

This scatter plot visualizes the relationship between "Sleep Hours" and "Exam Score". Looking at the plot, there doesn't appear to be a strong linear correlation between the number of sleep hours and the exam score. The points are quite scattered, indicating that while sleep is important, the amount of sleep hours alone doesn't seem to be a primary predictor of exam performance in this dataset. Other factors likely play a more significant role.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['attendance_percent'], df['exam_score'])
plt.xlabel('Attendance Percentage')
plt.ylabel('Exam Score')
plt.title('Attendance Percentage vs. Exam Score')
plt.grid(True)
```

This scatter plot shows the relationship between "Attendance Percentage" and "Exam Score". Similar to the "Sleep Hours" plot, there doesn't appear to be a very strong linear correlation between attendance percentage and exam score. While there might be a slight tendency for higher exam scores with higher attendance, the relationship is not as clear as the one between "Hours Studied" and "Exam Score". This suggests that while attendance is a factor, it might not be the most dominant one influencing exam performance in this dataset.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['previous_scores'], df['exam_score'])
plt.xlabel('Previous Scores')
plt.ylabel('Exam Score')
plt.title('Previous Scores vs. Exam Score')
plt.grid(True)
```

This scatter plot displays the relationship between "Previous Scores" and "Exam Score". Similar to the "Hours Studied" plot, there appears to be a positive correlation between these two variables. Students who had higher previous scores tend to have higher exam scores in this dataset. This suggests that past performance is a good indicator of future performance in this context.

```python
numerical_cols = ['hours_studied', 'sleep_hours', 'attendance_percent', 'previous_scores', 'exam_score']

for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    plt.hist(df[col], bins=15, edgecolor='black')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {col}')
    plt.show()
```
